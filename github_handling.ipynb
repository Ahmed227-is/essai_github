## <span style = "color:lime">**Importance de github**</span>

GitHub est important car il permet aux développeurs de **collaborer** sur des projets de manière efficace. Voici ses avantages simples :

1. **Sauvegarde du code** : GitHub conserve une copie de ton code en ligne, ce qui te permet de le retrouver même si tu perds ton ordinateur.

2. **Historique des modifications** : Il garde une trace de chaque changement effectué dans le code, facilitant ainsi le suivi des erreurs et la possibilité de revenir à une version précédente.

3. **Travail en équipe** : Plusieurs personnes peuvent travailler sur le même projet en même temps sans écraser le travail des autres, grâce à la gestion des branches.

4. **Partage et visibilité** : C'est un excellent outil pour montrer ton travail à d'autres développeurs ou employeurs, surtout si tu souhaites contribuer à des projets open source.

5. **Automatisation** : Avec **GitHub Actions**, tu peux automatiser des tâches comme les tests ou les déploiements du projet, ce qui te fait gagner du temps.

En résumé, GitHub facilite la **collaboration**, **l’organisation** et **l’automatisation** du travail des développeurs, tout en offrant un **historique clair** des modifications du code.

### <span style = "color:lime">**Les commandes de bases de git bash**</span>

Voici une liste des commandes **Git Bash** les plus importantes et fréquemment utilisées :

### 1. **Configuration Git**
   - `git config --global user.name "Ton Nom"` : Définit ton nom d'utilisateur global.
   - `git config --global user.email "tonemail@example.com"` : Définit ton email global.

### 2. **Commandes de base**
   - `git init` : Initialise un nouveau dépôt Git dans un dossier.
   - `git clone [url]` : Clone un dépôt existant depuis une URL vers ton ordinateur.
   - `git status` : Vérifie l'état du dépôt, montrant les modifications, les fichiers suivis et non suivis.
   - `git add [fichier]` : Ajoute un fichier spécifique pour le suivi.
   - `git add .` : Ajoute tous les fichiers modifiés dans le répertoire actuel.
   - `git commit -m "Message"` : Crée un commit avec un message descriptif.
   - `git push` : Envoie les modifications vers un dépôt distant (comme GitHub).
   - `git pull` : Récupère et intègre les modifications d'un dépôt distant.

### 3. **Gestion des branches**
   - `git branch` : Liste les branches existantes.
   - `git branch [nom_de_branche]` : Crée une nouvelle branche.
   - `git checkout [nom_de_branche]` : Bascule vers une autre branche.
   - `git merge [branche]` : Fusionne la branche spécifiée dans la branche actuelle.
   - `git branch -d [nom_de_branche]` : Supprime une branche localement.

### 4. **Annulation et réinitialisation**
   - `git reset [fichier]` : Désélectionne un fichier qui a été ajouté pour un commit.
   - `git reset --hard` : Réinitialise l'état du dépôt à celui du dernier commit.
   - `git revert [commit]` : Annule un commit spécifique tout en créant un nouveau commit.
   - `git stash` : Enregistre les modifications en cours sans les committer pour travailler sur autre chose.

### 5. **Travailler avec les dépôts distants**
   - `git remote add origin [url]` : Associe le dépôt local à un dépôt distant.
   - `git fetch` : Récupère les modifications du dépôt distant sans les fusionner.
   - `git remote -v` : Affiche les dépôts distants configurés.

### 6. **Affichage et logs**
   - `git log` : Affiche l'historique des commits.
   - `git diff` : Affiche les différences entre les fichiers suivis et non suivis.
   - `git show [commit]` : Montre les modifications d'un commit spécifique.

Ces commandes sont essentielles pour la gestion de projets en Git et facilitent le suivi des versions, le travail en équipe, et le développement structuré.

Le message que vous recevez signifie que Git a détecté une différence entre les fins de ligne (ou **line endings**) dans votre fichier. Voici ce qui se passe :

- **LF (Line Feed)** est le caractère utilisé comme fin de ligne sous Linux et macOS.
- **CRLF (Carriage Return + Line Feed)** est le caractère utilisé comme fin de ligne sous Windows.

Git détecte que dans le fichier `github_handling.ipynb`, les fins de ligne actuelles sont en format **LF**. Cependant, Git est configuré pour remplacer ces fins de ligne par **CRLF** lorsque vous modifierez à nouveau ce fichier sur votre système Windows.

### Comment gérer cela ?
1. **Laisser Git gérer les fins de ligne automatiquement** : Git essaie de normaliser les fins de ligne en fonction du système d'exploitation. Il est souvent configuré pour convertir les fins de ligne en **LF** sur le dépôt distant et en **CRLF** sur Windows pour que le fichier reste lisible localement.

2. **Configurer Git pour éviter les avertissements** :
   Si vous souhaitez éviter cet avertissement à l'avenir, vous pouvez configurer Git pour gérer les fins de ligne correctement selon votre besoin :

   - Si vous voulez que Git convertisse automatiquement les fins de ligne en **LF** dans le dépôt et en **CRLF** dans votre système local, exécutez :
     ```bash
     git config --global core.autocrlf true
     ```

   - Si vous voulez que Git ne modifie jamais les fins de ligne, utilisez :
     ```bash
     git config --global core.autocrlf false
     ```

   - Si vous préférez que Git conserve les fins de ligne en **LF** partout (idéal pour les projets multi-plateformes), vous pouvez utiliser :
     ```bash
     git config --global core.eol lf
     git config --global core.autocrlf input
     ```

Cela devrait résoudre les avertissements liés aux fins de ligne lors de l'ajout de fichiers à votre dépôt.
